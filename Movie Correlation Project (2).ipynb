```python
# import libraries

import pandas as pd 
import numpy as np 
import seaborn as sns

import matplotlib.pyplot as plt 
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure 

%matplotlib inline 
plt.rcParams['figure.figsize'] = (12,8) #adjusts the configuration of the plot we will create

# Read in the data
df = pd.read_csv(r'C:\Data\movies.csv') 
```

```python
# Now let's take a look at the data
df.head()
```

```python
# Let's see if there is any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,  round(pct_missing*100)))
```

```python
# Data Types for our columns
print(df.dtypes)
```

```python
df.head()
```

```python

# change data type of columns

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
# Fill missing values before converting to int64 and trying again

df['budget'] = df['budget'].fillna(0).astype('int64')
df['gross'] = df['gross'].fillna(0).astype('int64')
```

```python
df.head()
```

```python
#Create correct year column

df['yearcorrect']=df['released'].astype(str).str[:4]
df.head()
```

```python
df= df.sort_values(by='gross',inplace=False,ascending=False)
```

```python
#show scrolling on display data

pd.set_option('display.max_rows', None)
```

```python
# Drop any duplicates 
#show distinct values

df['company'].drop_duplicates().sort_values(ascending=False)
df.head()
```

```python
df.head()
```

```python
#Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earning')
plt.xlabel('Gross Earning')
plt.ylabel('Budget for Film')
plt.show()
```

```python
df.head()
```

```python
#regression plot using Seaborn
#plot the budget vs gross using seaborn

sns.regplot(x='budget', y='gross',data=df, scatter_kws={'color': 'red'},line_kws={'color': 'blue'})
plt.show()
```

```python
#visualization

correlation_matrix =df.corr(numeric_only=True, method='pearson')

sns.heatmap(correlation_matrix, annot= True )
plt.title('Correlation Matrix For Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
 #dark= lowcorrelation
#light= highcorrelation
```

```python
df_numerized = df.copy()

for col_name in df_numerized.columns:
    if df_numerized[col_name].dtype == 'object':
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()

                                         
```

```python
df.head()
```

```python
#visualization

correlation_matrix =df_numerized.corr(numeric_only=True, method='pearson')

sns.heatmap(correlation_matrix, annot= True )
plt.title('Correlation Matrix For Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
 #dark= lowcorrelation
#light= highcorrelation
```

```python
df_numerized.corr()
```

```python
#unstack

correlation_mat =df_numerized.corr()
corr_pairs =correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs =corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr= sorted_pairs[(sorted_pairs) >0.5]

high_corr
```
